```python
import arcpy

fc = r"Csat_update_kodkeke"   # your feature class (POLYLINE)

sr  = arcpy.Describe(fc).spatialReference
tol = getattr(sr, "XYTolerance", None)
tol = float(tol) if tol not in (None, 0) else 1e-6      # tiny fallback in map units

def is_closed(g):
    """True if every part of a polyline starts and ends at the same place (within tol)."""
    if g is None or g.pointCount < 2:
        return False
    for i in range(g.partCount):
        arr = g.getPart(i)                               # arcpy.Array of Points/None
        pts = [p for p in arr if p]                      # strip None separators
        if len(pts) < 2:
            return False
        p0 = arcpy.PointGeometry(pts[0], sr)
        p1 = arcpy.PointGeometry(pts[-1], sr)
        if p0.distanceTo(p1) > tol:                      # tolerance-aware comparison
            return False
    return True

# --- delete opens in-place (keeps every attribute, incl. `date`) ---
kept, dropped = 0, 0
with arcpy.da.UpdateCursor(fc, ["OID@", "SHAPE@"]) as cur:
    for oid, g in cur:
        if not is_closed(g):
            cur.deleteRow()
            dropped += 1
        else:
            kept += 1

print(f"Kept closed polylines: {kept}")
print(f"Removed open polylines: {dropped}")
```
